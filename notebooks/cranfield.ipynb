### Loading the documents

Next, we will define a class named `Document` that will represent a preprocessed document from the Cranfield collection. Tokenization and preprocessing of the `title` and `body` attributes of the individual documents as well as the creative use of the `authors`, `bibliography`, and `title` attributes is left to your imagination and craftsmanship.

```python
from pv211_utils.cranfield.entities import CranfieldDocumentBase

class Document(CranfieldDocumentBase):
    """
    A preprocessed Cranfield collection document.

    Parameters
    ----------
    document_id : str
        A unique identifier of the document.
    authors : list of str
        A unique identifiers of the authors of the document.
    bibliography : str
        The bibliographical entry for the document.
    title : str
        The title of the document.
    body : str
        The abstract of the document.

    """
    def __init__(self, document_id: str, authors: str, bibliography: str, title: str, body: str):
        super().__init__(document_id, authors, bibliography, title, body)

    def __str__(self):
        return f"{self.title} {self.body}"
```

## Implementation of your information retrieval system

You can try the [preprocessing][1] and [systems][2] that are [available in our library][1], but feel free to implement your own.

 [1]: https://github.com/MIR-MU/pv211-utils/tree/main/pv211_utils/preprocessing

```python
from pv211_utils.systems import BM25PlusSystem
from pv211_utils.preprocessing import DocPreprocessingBase
from gensim.utils import simple_preprocess, deaccent
import nltk
nltk.download('stopwords', quiet=True)
from nltk.stem import PorterStemmer

class BetterPreprocessing(DocPreprocessingBase):
    def __init__(self, min_len=2, max_len=15):
        self.min_len = min_len
        self.max_len = max_len
        self.stopwords = set(stopwords.words('english'))
        self.stopwords.update([
            "paper", "study", "result", "method",
            "using", "used", "show"
        ])
        self.stemmer = PorterStemmer()

    def __call__(self, doc: str):
        tokens = simple_preprocess(doc, deacc=True)

        processed = []
        for word in tokens:
            if (
                word not in self.stopwords and
                self.min_len <= len(word) <= self.max_len and
                word.isalpha()
            ):
                word = self.stemmer.stem(word)
                processed.append(word)

        bigrams = [
            processed[i] + "_" + processed[i+1]
            for i in range(len(processed) - 1)
            if processed[i] not in self.stopwords and processed[i+1] not in self.stopwords
        ]

        return processed + bigrams

class BM25PRFSystem:
    def __init__(self, base_system, preprocessing, top_k=10, expand_terms=15):
        self.base = base_system
        self.preprocessing = preprocessing
        self.top_k = top_k
        self.expand_terms = expand_terms

    def expand_query(self, query_tokens, top_docs):
        term_scores = {}

        for doc_id, doc in enumerate(top_docs[:self.top_k]):
            tokens = self.preprocessing(str(doc))

            weight = 1.0 / (doc_id + 1)

            for t in tokens:
                if t in query_tokens:
                    continue

                term_scores[t] = term_scores.get(t, 0) + weight

        max_score = max(term_scores.values()) if term_scores else 1
        for t in term_scores:
            term_scores[t] /= max_score

        sorted_terms = sorted(term_scores.items(), key=lambda x: x[1], reverse=True)

        expansion = []
        for t, _ in sorted_terms:
            if len(t) <= 3 or "_" in t:
                continue
            expansion.append(t)
            if len(expansion) >= self.expand_terms:
                break

        return expansion

    def search(self, query):
        query_tokens = self.preprocessing(str(query))
        initial_results = list(self.base.search(query))
        expanded_tokens = self.expand_query(query_tokens, initial_results)
        expanded_tokens = query_tokens * 4 + expanded_tokens
        scores = self.base.bm25.get_scores(expanded_tokens)
        ranked = scores.argsort()[::-1]

        for doc_id in ranked:
            yield self.base.index[doc_id]

preprocessing = BetterPreprocessing()
base_system = BM25PlusSystem(documents, preprocessing, k1=1.8, b=0.6)
system = BM25PRFSystem(base_system, preprocessing)
```
